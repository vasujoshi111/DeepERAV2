```python
import cv2, os
src = r"./output"
image_paths = [os.path.join(src, j) for j in os.listdir(src)]
```

```python
num_images = len(image_paths)
fig, axes = plt.subplots(2, 5, figsize=(25, 6))

# Loop through the images and their corresponding outputs
for i, (image_path) in enumerate(image_paths):
    # Load the image using OpenCV
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Display the image
    ax = axes[i // 5, i % 5]
    ax.imshow(image)
    ax.axis('off')
    ax.set_title(f'Image {i+1}')

plt.tight_layout()
plt.show()
```
